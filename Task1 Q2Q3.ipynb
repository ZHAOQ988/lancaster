```python
# import libraries
import pandas as pd
import statsmodels.api as sm
import csv
import statsmodels.formula.api as smf
import numpy as np

#For Estimation of Beta for Johnson and Johnson

jnj = pd.read_csv (r'/Users/celine51000/Desktop/JNJ.csv', parse_dates=True, index_col='Date',)
print (jnj)



```

```python
import numpy as np
sp_500 = pd.read_csv(r'/Users/celine51000/Desktop/^GSPC.csv', parse_dates=True, index_col='Date',)
print (sp_500)
```

```python
# joining the closing prices of the two datasets 
monthly_prices = pd.concat([jnj['Close'], sp_500['Close']], axis=1)
monthly_prices.columns = ['JNJ','^GSPC']
```

```python
# check the head of the dataframe
print(monthly_prices.head())
```

```python
# calculate monthly returns
monthly_returns = monthly_prices.pct_change(1)
clean_monthly_returns = monthly_returns.dropna(axis=0)  # drop first missing row
print(clean_monthly_returns.head())
```

```python
# split dependent and independent variable
X = clean_monthly_returns['^GSPC']
y = clean_monthly_returns['JNJ']

# Add a constant to the independent value
X1 = sm.add_constant(X)

# make regression model 
model = sm.OLS(y, X1)

# fit model and print results
results = model.fit()
print(results.summary())
```
